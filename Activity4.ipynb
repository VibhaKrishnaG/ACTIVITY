# 3) Compare 'tenure' with 'Total Charges'

```python
sns.scatterplot(x=telco['tenure'],y=telco['TotalCharges'],data=telco,hue='SeniorCitizen')
plt.title('Comparison of Tenure with Total Charges')

plt.grid()
plt.tight_layout()
plt.show()
```

As tenure increased the total charges also increased.

# 4) Find out which contract is preffered by the senior citizen.

```python
sns.countplot(x=telco[telco['SeniorCitizen']==1]['Contract'])
plt.grid()
plt.tight_layout()
plt.show()
```

The contract preffered by the senior citizens is Month-to-month.
